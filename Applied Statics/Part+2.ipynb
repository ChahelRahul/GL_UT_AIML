```python
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
Team = pd.read_csv('DS - Part2 - Basketball.csv',index_col=0)
```

```python
Team
```

```python
Team.info()
```

```python
Team['TournamentChampion'] = Team['TournamentChampion'].replace('-', 0)
Team['Runner-up'] = Team['Runner-up'].replace('-', 0)
Team['TeamLaunch'] = Team['TeamLaunch'].str[0:4]
Team.iloc[60]=Team.iloc[60].replace('-', 0)
```

```python
Team[Team.columns]=Team[Team.columns].apply(pd.to_numeric)
```

```python
Team.info()
```

```python
Team
```

```python
Team.describe()
```

```python
#univarient analysis


plt.figure(figsize=(20,6))
plt.subplot(1, 3, 1)
plt.title('Tournament')
sns.distplot(Team['Tournament'],color='green')

# subplot 2
plt.subplot(1, 3, 2)
plt.title('Score')
sns.distplot(Team['Score'],color='blue')

# subplot 3l
plt.subplot(1, 3, 3)
plt.title('WonGames')
sns.distplot(Team['WonGames'],color='red')

plt.figure(figsize=(20,6))
plt.subplot(1, 3, 1)
plt.title('Tournament')
sns.boxplot(Team['Tournament'],orient='vertical',color='green')

# subplot 2
plt.subplot(1, 3, 2)
plt.title('Score')
sns.boxplot(Team['Score'],orient='vertical',color='blue')

# subplot 3
plt.subplot(1, 3, 3)
plt.title('WonGames')
sns.boxplot(Team['WonGames'],orient='vertical',color='red')


plt.show()
```
